```python
#GROUPING:(.groupby())
import pandas as pd
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  
print(df)
df.groupby("Department")["Salary"].mean()#Group by Department, then calculate average Salary for each group
#other methods
df.groupby("Team")["Salary"].mean()     # Average per team
df.groupby("Team")["Salary"].sum()      # Total score
df.groupby("Team")["Salary"].count()    # How many entries
df.groupby("Team")["Salary"].min()      # min of the team 
df.groupby("Team")["Salary"].max()      #max of a team
#To group by multiple columns:
df.groupby(["Team", "Gender"])["Salary"].mean()
#AGGREGATION:(multiple methods/functions at once)
df.groupby("Department")["Salary"].agg(["mean","max","min","count","sum"])
df.groupby("Department")["Salary"].agg(avg="mean",no_of_people="count") #name your columns in fuction 
df.groupby("Department").agg({"Salary":"mean","Age":"max"})#aggregate particular columns separately
#aggregate returns single value per group, used to give summary(like mean)
#TRANSFORM:(same shape as original,adds new column based on group)
df["Team Avg"] = df.groupby("Team")["Salary"].transform("mean")
#FILTER:(it returns a subset of rows , used to keep/discard whole groups)
df.groupby("Team").filter(lambda x: x["Salary"].mean() > 80)
```

```python
#MERGING AND JOINING DATA:
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})

departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
#MERGE LIKE SQL:(inner, left, right, outer)
pd.merge(employees,departments,on="DeptID")#excludes uncommon("inner")
pd.merge(employees,departments,on="DeptID",how="left")#will keep all employees , nan at nomatch
pd.merge(employees,departments,on="DeptID",how="right")#will keep all departments
pd.merge(employees,departments,on="DeptID",how="outer")#will show both left and right
#CONCATENATING DATAFRAMES:(used for stacking dataframes)
df1 = pd.DataFrame({"Name": ["Alice", "Bob"]})
df2 = pd.DataFrame({"Name": ["Charlie", "David"]})
pd.concat([df1, df2])#adds horizontally
pd.concat([df1,df2],axis=1)#adds vertically
```
